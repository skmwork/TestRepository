```python
import numpy as np
import matplotlib.pyplot as plt

N = 10000
T = np.linspace(0, 1000, N+1, dtype=np.float32)

a=3
b=10
X = np.cos(a*T)
Y = np.sin(b*T)

fig = plt.figure()
ax = plt.subplot()
ax.plot(X, Y, 'g-')
fig.savefig(f'liss_{a}_{b}.png')
```

фигуры лиссажу от 1 до 9
прямая методом наименьшиъ квадратов с индексами от 1 до 9 включительно

```python
x = np.arange(-5, 6, 0.3)
```

```python
y = 0.5*x + 4 + np.random.rand(*x.shape)/10
```

```python
plt.plot(x,y,'go')
```

```python
N = x.shape[0]
```

```python
N
```

```python
a1 = N*sum(x*y)
```

```python
a2 = sum(x)*sum(y)
```

```python
a3 = N*sum(x**2)
```

```python
a4 = sum(x)**2
```

```python
a = (a1 - a2)/(a3 - a4)
```

```python
b = (sum(y) - a*sum(x))/N
```

```python
y1 = a*X + b
```

```python
plt.plot(x,y,x,y1)
```
